```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("yankihue/gpt2-tr-detoxified-final")

model = AutoModelForCausalLM.from_pretrained("yankihue/gpt2-tr-detoxified-final")



tokenizer.eos_token_id = model.config.eos_token_id
tokenizer.pad_token = tokenizer.eos_token
```

```python
from datasets import load_dataset

dataset = load_dataset("yankihue/tweets-turkish",split="train")
```

```python
import pandas as pd
```

```python
dataset
```

```python
def filter_fn(sample):
        is_toxic = sample["Tip"]
        return is_toxic == "Pozitif"  

dataset = dataset.filter(filter_fn, batched=False)
```

```python
df = dataset.to_pandas()
```

```python
df
```

```python
df_sampled = df.sample(n=6000)
```

```python
df_sampled
```

```python
import numpy as np
```

```python
def random_substring(s):
    if len(s) <= 10:
        return s
    elif 10 < len(s) < 40:
        end = np.random.randint(10, len(s))
        return s[:end]
    else:
        end = np.random.randint(10, 41)
        return s[:end]

```

```python
df_sampled['paylaşım_substring'] = df_sampled['Paylaşım'].apply(random_substring)
```

```python
gen_kwargs = {"min_length": -1, "top_k": 0.0, "top_p": 1.0, "do_sample": True, "pad_token_id": tokenizer.eos_token_id}
```

```python
from transformers import pipeline
```

```python
import torch
```

```python
outputs = list()
```

```python
from trl.core import LengthSampler

output_min_length = 4
output_max_length = 16
output_length_sampler = LengthSampler(output_min_length, output_max_length)

gen_len=output_length_sampler()
```

```python
for i in range(6000):
  input_text = df_sampled.iloc[i]['paylaşım_substring']
  input_ids = tokenizer.encode(input_text, return_tensors='pt')

  # generate output from the model
  output = model.generate(input_ids,max_new_tokens=gen_len, **gen_kwargs)

  decoded_output = tokenizer.decode(output[0], skip_special_tokens=True)
  outputs.append(decoded_output)
```

```python
len(outputs)
```

```python
df2 = pd.DataFrame({"query": df_sampled['paylaşım_substring'], "output": outputs },columns=["query","output"])
```

```python
df2
```

```python
df2['output'] = df2.apply(lambda row: row['output'].replace(row['query'], ''), axis=1)
```

```python
df2.to_csv("detoxified 6000.csv",index = False)
```
