```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("redrussianarmy/gpt2-turkish-cased")

model = AutoModelForCausalLM.from_pretrained("redrussianarmy/gpt2-turkish-cased")



tokenizer.eos_token_id = model.config.eos_token_id
tokenizer.pad_token = tokenizer.eos_token
```

```python
from datasets import load_dataset

dataset = load_dataset("yankihue/tweets-turkish",split="train")
```

```python
df_sampled = df.sample(n=2000)
```

```python
df_sampled
```

```python
import numpy as np
```

```python
def random_substring(s):
    if len(s) <= 10:
        return s
    elif 10 < len(s) < 40:
        end = np.random.randint(10, len(s))
        return s[:end]
    else:
        end = np.random.randint(10, 41)
        return s[:end]

```

```python
df_sampled['paylaşım_substring'] = df_sampled['Paylaşım'].apply(random_substring)
```

```python
gen_kwargs = {"min_length": -1, "top_k": 0.0, "top_p": 1.0, "do_sample": True, "pad_token_id": tokenizer.eos_token_id}
```

```python
from transformers import pipeline
```

```python
import torch
```

```python
outputs1 = list()
outputs2 = list()
```

```python
from trl.core import LengthSampler

output_min_length = 4
output_max_length = 16
output_length_sampler = LengthSampler(output_min_length, output_max_length)

gen_len=output_length_sampler()
```

```python
for i in range(2000):
  input_text = df_sampled.iloc[i]['paylaşım_substring']
  input_ids = tokenizer.encode(input_text, return_tensors='pt')

  # generate output from the model
  output1 = model.generate(input_ids,max_new_tokens=gen_len, **gen_kwargs)
  output2 = model.generate(input_ids,max_new_tokens=gen_len, **gen_kwargs)


  decoded_output1 = tokenizer.decode(output1[0], skip_special_tokens=True)
  decoded_output2 = tokenizer.decode(output2[0], skip_special_tokens=True)

  outputs1.append(decoded_output1)
  outputs2.append(decoded_output2)
```

```python
df2 = pd.DataFrame({"prompt": df_sampled['paylaşım_substring'], "chosen": outputs1, "rejected":outputs2 },columns=["prompt","chosen","rejected"])
```

```python
df2
```

```python
df2['chosen'] = df2.apply(lambda row: row['chosen'].replace(row['prompt'], ''), axis=1)
df2['rejected'] = df2.apply(lambda row: row['rejected'].replace(row['prompt'], ''), axis=1)

```

```python
df2.to_csv("unranked_outputs_full.csv",index = False)
```
